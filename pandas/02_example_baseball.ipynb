```python
import numpy as np
import pandas as pd
```

```python
baseball = pd.read_excel("./data/2019_baseball_hitter20.xlsx", sheet_name=0)
baseball
```

```python
baseball.head()
```

```python
data1 = baseball.sort_values(by=['홈런'], ascending=False)
data1.head()
```

```python
del data1['순번']
data1.head()
```

```python
data1.reset_index(drop=True, inplace=False)
```

```python
data1.loc[:,['이름','홈런']].head()
```


#  그래프 그리기

```python
import matplotlib.pyplot as plt
```

```python
data1.이름
```

```python
names = data1.이름.tolist()
names
```

```python
trimmed_names = []
for name in names:
    trimmed_names.append(name.strip())

trimmed_names
```

```python
data1.이름 = trimmed_names
```

```python
data1.이름.tolist()
```

```python
homerun = data1.홈런.tolist()
homerun
```

```python
import matplotlib.font_manager as fm
import platform
font_path = 'C:/Windows/Fonts/gulim.ttc'
```

```python
font_name = fm.FontProperties(fname=font_path).get_name()
plt.rcParams['font.family'] = font_name
plt.rcParams['font.size'] = 12
```

```python
plt.figure(figsize=(20,4))
plt.plot(trimmed_names, homerun)
plt.show()
```

```python
plt.figure(figsize=(14,4))
plt.bar(trimmed_names, homerun, color='#0000ff')
plt.xlabel('선수이름')
plt.xticks(rotation=45)
plt.ylabel('홈런수')
plt.title('2018년 프로야구 홈런타자 순위')
plt.show()
```
