```python
import cv2         #openCV module
import numpy as np

img = np.zeros(shape = (512, 512, 3), dtype = np.uint8) + 255      # 2 dimesional image array all value 255 unsigned int 8bit type
```

```python
pt1 = 100, 100
pt2 = 400, 400
```

```python
rec = cv2.rectangle(img, pt1, pt2, (0, 255, 0), 2)
```

```python
import matplotlib.pyplot as plt
```

d## 

```python
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) #BGR type을 RGB type으로 convert

plt.imshow(img2)
plt.show()
```

```python
cv2.line(img, (30,30), (500,30), (255, 0, 0), 5)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```

```python
cv2.circle(img, (250, 250), 150, (0, 0, 255), -1)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img2)
plt.show()
```
