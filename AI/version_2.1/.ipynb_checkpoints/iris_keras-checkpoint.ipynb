```python
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam

import tensorflow as tf
import numpy as np
```

```python
data = datasets.load_iris()   #  붓꽃데이터 불러옮
```

```python
X = data.data
y = data.target
y = y.reshape(-1, 1)  # column vector 로 변환
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# 8 : 2 로 trainig : test 데이터 분할
```

```python
scaler = StandardScaler()
scaler.fit(X_train)   # Regularization 위해서 평균, 표준편차 계산
X_train = scaler.transform(X_train)
```

```python
enc = OneHotEncoder()
enc.fit(y)   # target 종류를 파악해서 다중 분류 데이터로 손질
```

```python
y_train_onehot = enc.transform(y_train).toarray()
y_train_onehot = np.array(y_train_onehot, dtype="float32")
# nd.array로 반환되는데 내부 element type을 float32 로 변환
```

```python
y_test_onehot = enc.transform(y_test).toarray()
y_test_onehot = np.array(y_test_onehot, dtype="float32")
```

```python
model = Sequential()

model.add(Dense(16, input_dim = 4, activation="relu"))   # 0~1 사이 값으로 변환
model.add(Dense(3, activation="softmax"))   # 전체 합이 1이 되도록 확률 모댈링
model.summary()
```

```python
model.compile(loss="categorical_crossentropy", optimizer=Adam(lr=0.1), metrics=["acc"])
model.fit(X_train, y_train_onehot, epochs=1000)
```

```python
X_test = scaler.transform(X_test)    # 훈련 데이터의 평균, 표준편차로 정규화
predict = model.predict(X_test)
predict01 = np.argmax(predict, 1)   # 확률이 최대인 output value를 부여
predict01
```

```python
target = np.argmax(y_test_onehot, 1)
target
```

```python
predict02 = (predict01 == target)
predict02
```

```python
np.sum(predict02)
acc = np.sum(predict02) / len(predict02)
acc
```
