```python
import numpy as np
import tensorflow as tf
tf.enable_eager_execution()
```

```python
X = np.array([
    [0., 0.],
    [0., 1.],
    [1., 0.],
    [1., 1.]            
], dtype="float32")

y = np.array([
    [0.],
    [0.],
    [0.],
    [1.]
], dtype="float32")
```

```python
y
```

# hypothesis = w1 x1 + w2 x2 + b 
# initial value 
# w1 = 1, w2 = 2, b =3

```python
w = np.array([[1], [2]], dtype="float32")

b = np.array([3], dtype="float32")
```

```python
w
```

```python
b
```

```python
np.dot(X,w)
```

```python
np.dot(X,w) + b
```

```python
def sigmoid(h):
    return 1 / (1 + np.exp(-h))
```

```python
hypothesis = np.dot(X,w) + b
hypothesis
```

```python
hypothesis = sigmoid(hypothesis)
hypothesis
```

```python
error = hypothesis - y
error
```

```python
X.T[0]
```

```python
X_col_2d1 = X.T[0].reshape(1, -1)
```

```python
w1_grad = 1 / 4 * np.sum(np.dot(X_col_2d1, error))
print("w_grad:", w1_grad)
```

```python
X.T[1]
```

```python
X_col_2d2 = X.T[1].reshape(1, -1)
```

```python
w2_grad = 1 / 4 * np.sum(np.dot(X_col_2d2, error))
print("w_grad:", w2_grad)
```

```python
learning_rate = 0.1

for i in range(500):
    hypothesis = sigmoid(np.dot(X, w) + b)

    cost = -1/4 * (y*np.log(hypothesis) + (1 - y) * np.log(1 - hypothesis))

    X_col_2d0 = X.T[0].reshape(1, -1)
    w0_grad = 1/4 * np.sum(np.dot(X_col_2d0, (hypothesis - y)))
    w0_new = w[0]-learning_rate * w0_grad
    
    X_col_2d1 = X.T[1].reshape(1, -1)
    w1_grad = 1/4 * np.sum(np.dot(X_col_2d1, (hypothesis - y)))
    w1_new = w[1]-learning_rate * w1_grad
    
    b_grad = 1/4 * np.sum(error)
    
    b_new = b - learning_rate * b_grad
    
    w[0] = w0_new
    w[1] = w1_new
    b = b_new
    
    print("epoch[{}]  w[0]= {}, w[1]= {}, b= {}".format(i+1, w[0], w[1], b))
```
