```python
import cv2
import numpy as np
from math import cos, radians

width, height = 512, 512
x, y, R = 256, 256, 50
direction = 0

top_lim = R
bot_lim = int(R * cos(radians(36)))
left_lim = int(R * cos(radians(18)))

mouse_img = np.zeros((width, height, 3), np.uint8) + 255

def onMouse(event, x, y, flags, param):
    global mouse_img
    
    if event == cv2.EVENT_LBUTTONDOWN:    
        text = "Hello OpenCV"
        org = (x, y)
        font = cv2.FONT_HERSHEY_SIMPLEX
        cv2.putText(mouse_img, text, org, font, 1, (255, 0, 0), 2)
    
    elif event == cv2.EVENT_RBUTTONDOWN:
        cv2.rectangle(mouse_img, (x -75, y-75), (x + 75, y + 75), (0, 0, 255))
    
    elif event == cv2.EVENT_LBUTTONDBLCLK:
        mouse_img = np.zeros(param[0].shape, np.uint8) + 255
    
while True:
    key = cv2.waitKeyEx(30)
    if key == 0x1B:
        break;
    
    elif key == 0x270000:
        direction = 0
    elif key == 0x280000:
        direction = 1
    elif key == 0x250000:
        direction = 2
    elif key == 0x260000:
        direction = 3
    
    if direction == 0:
        x += 10
    elif direction == 1:
        y += 10
    elif direction == 2:
        x -= 10
    else:
        y -= 10        
    
    if x < left_lim:
        x = left_lim
        direction = 0
    if x > width - left_lim:
        x = width - left_lim
        direction = 2
    if y < top_lim:
        y = top_lim
        direction = 1
    if y > height - bot_lim:
        y = height - bot_lim
        direction = 3
    
    
    img = mouse_img.copy()
    pts = cv2.ellipse2Poly((x,y), (50,50), 0, -90, 270, 72)
    cv2.polylines(img, [pts], True, (0, 0, 255), 2)
    cv2.imshow('img', img)
    
    cv2.setMouseCallback('img', onMouse, [img])
    
cv2.destroyAllWindows()
```
